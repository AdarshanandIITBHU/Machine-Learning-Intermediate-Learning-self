```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('/content/drive/My Drive/Colab Notebooks/datasets/IPL Matches 2008-2020.csv',converters={'team1':rename})
```
